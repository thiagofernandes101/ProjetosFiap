```python
import pandas as pd
import sklearn as sk
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_frame = pd.read_csv('./DropOutInaugural.csv')
data_frame
```

```python
data_frame.info()
data_frame.describe()
```

```python
# drop null values
data_frame.dropna(inplace=True)
data_frame.describe()
```

```python
# drop duplicates
data_frame.drop_duplicates(inplace=True)
data_frame.describe()
# drop irrelevant columns
```

```python
# Lista das variáveis categóricas
categorias = [
    'Marital status', 'Course', 'Daytime/evening attendance', 'Previous qualification',
    'Nacionality', 'Mother\'s qualification', 'Father\'s qualification', 'Mother\'s occupation',
    'Father\'s occupation', 'Displaced', 'Educational special needs', 'Debtor', 'Tuition fees up to date',
    'Gender', 'Scholarship holder'
]

# Plot de gráficos de barras para cada variável categórica em relação ao Target
for var in categorias:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=var, hue='Target', data=data_frame, palette='Set2')
    plt.title(f'Dropouts e Graduates por {var}')
    plt.xticks(rotation=45)
    plt.show()
```

```python
X = data_frame.drop('Target', axis=1)  # Todas as colunas, exceto 'Target'
y = data_frame['Target']  # Apenas a coluna 'Target'
```

```python
from sklearn.model_selection import train_test_split

# Dividir os dados entre características (X) e rótulo (y)
X = data_frame.drop('Target', axis=1)  # Características
y = data_frame['Target']  # Rótulo

# Dividir em treino (70%) e teste (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)

# Verificar as dimensões para garantir que a divisão foi realizada corretamente
print(f"Tamanho do conjunto de treino (X_train): {X_train.shape}")
print(f"Tamanho do conjunto de teste (X_test): {X_test.shape}")
```

# Regressão Logística

```python
# treinamento de modelo de classificação a sua escolha (um ou mais)
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

# Supondo que já temos os dados X_train, X_test, y_train, y_test definidos
# Dividindo os dados em treino e teste (se já não foi feito anteriormente)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)

# --- Modelo 1: Regressão Logística ---
# Inicializando o modelo de Regressão Logística
logreg = LogisticRegression(max_iter=1000)

# Treinando o modelo com os dados de treino
logreg.fit(X_train, y_train)

# Fazendo previsões no conjunto de teste
y_pred_logreg = logreg.predict(X_test)

# Avaliando o desempenho (acurácia)
acc_logreg = accuracy_score(y_test, y_pred_logreg)
print(f"Acurácia do modelo de Regressão Logística: {acc_logreg:.4f}")
```

**1. Regressão Logística:**

- Definição: A Regressão Logística é um modelo de classificação que usa uma função sigmoide para modelar a probabilidade de uma classe em um problema de classificação binária. Ela calcula uma função linear dos atributos (variáveis independentes) e aplica a função sigmoide (ou logística) para produzir um valor entre 0 e 1. Esse valor é interpretado como a probabilidade de uma amostra pertencer à classe positiva.

- Explicação do Resultado: Se você obteve uma acurácia alta, isso significa que a separação entre as duas classes (dropouts e graduates) é bastante linear, ou seja, as variáveis explicativas têm uma relação clara com o rótulo.

- Caso a acurácia seja baixa, isso pode indicar que a relação entre as variáveis e a classe não é linear o suficiente para que a Regressão Logística seja eficaz.

# Random Forest

```python
# --- Modelo 2: Random Forest ---
# Inicializando o modelo de Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)

# Treinando o modelo com os dados de treino
rf.fit(X_train, y_train)

# Fazendo previsões no conjunto de teste
y_pred_rf = rf.predict(X_test)

# Avaliando o desempenho (acurácia)
acc_rf = accuracy_score(y_test, y_pred_rf)
print(f"Acurácia do modelo de Random Forest: {acc_rf:.4f}")
```

**2. Random Forest:**

- Definição: O Random Forest é um ensemble de árvores de decisão. Cada árvore é construída a partir de uma amostra aleatória dos dados e usa uma seleção aleatória das características para dividir os dados. A saída do modelo é determinada pela média (para regressão) ou pela votação das árvores (para classificação). O Random Forest reduz o risco de overfitting, pois combina várias árvores, tornando o modelo mais robusto.

- Explicação do Resultado: Se a acurácia do Random Forest for maior que a da Regressão Logística, pode ser um sinal de que o problema envolve interações complexas entre as variáveis e que a separação entre as classes não é tão simples ou linear. A Random Forest pode capturar essas interações melhor do que a Regressão Logística.

- Acurácia baixa no Random Forest: Isso pode ocorrer se as árvores estiverem subajustadas ou o número de árvores não for suficiente.

# Support Vector Machine (SVM)

```python
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# --- Modelo 3: Support Vector Machine (SVM) ---
# Inicializando o modelo SVM
svm = SVC(kernel='linear', random_state=42)

# Treinando o modelo com os dados de treino
svm.fit(X_train, y_train)

# Fazendo previsões no conjunto de teste
y_pred_svm = svm.predict(X_test)

# Avaliando o desempenho (acurácia)
acc_svm = accuracy_score(y_test, y_pred_svm)
print(f"Acurácia do modelo SVM: {acc_svm:.4f}")
```

3. Support Vector Machine (SVM):
- Definição: O SVM é um algoritmo de classificação que encontra o hiperplano de separação que maximiza a margem entre as classes. Ou seja, ele tenta encontrar a melhor linha (ou plano, ou hiperplano em dimensões mais altas) que separa as classes de maneira mais clara possível. O SVM é eficiente tanto para problemas lineares quanto não lineares (quando usa kernels).

- Explicação do Resultado: Se a acurácia do SVM for boa, isso pode indicar que as classes são separáveis por um hiperplano, ou seja, o problema é linearmente separável (pelo menos parcialmente).

- SVM com baixo desempenho: Pode ser que o kernel linear não tenha sido suficiente para separar as classes, ou seja, você pode precisar de um kernel não linear (como rbf ou poly), ou talvez a distribuição dos dados não seja tão clara para o SVM.

# K-Nearest Neighbors (KNN)

```python
# --- Modelo 4: K-Nearest Neighbors (KNN) ---
# Inicializando o modelo KNN
knn = KNeighborsClassifier(n_neighbors=5)

# Treinando o modelo com os dados de treino
knn.fit(X_train, y_train)

# Fazendo previsões no conjunto de teste
y_pred_knn = knn.predict(X_test)

# Avaliando o desempenho (acurácia)
acc_knn = accuracy_score(y_test, y_pred_knn)
print(f"Acurácia do modelo KNN: {acc_knn:.4f}")
```

4. K-Nearest Neighbors (KNN):

- Definição: O KNN é um método baseado em instâncias, onde a classe de um ponto de dados é atribuída com base nas classes dos K vizinhos mais próximos. O KNN não faz suposições sobre a distribuição dos dados e é muito simples. Ele apenas compara a proximidade entre os pontos para determinar as classes.

- Explicação do Resultado: Se a acurácia do KNN for boa, isso indica que as classes podem ser bem separadas com base na proximidade espacial dos dados. Em problemas simples ou com poucas dimensões, o KNN pode ter um bom desempenho.

- Acurácia baixa: O KNN pode ser muito sensível à escala dos dados (as variáveis precisam estar na mesma escala), e o desempenho pode ser afetado por isso. Além disso, a escolha do número de vizinhos (K) é crucial para o desempenho: um valor muito pequeno de K pode resultar em um modelo sensível ao ruído, enquanto um valor grande pode levar a um modelo mais generalista. Também pode ser necessário mais tempo computacional para grandes bases de dados.

# Resumo das Comparações:

- Se a Acurácia do SVM for melhor: Isso indica que existe uma separação clara e precisa entre as classes, e o SVM conseguiu encontrar esse limite de forma eficiente.

- Se a Acurácia do KNN for melhor: Isso sugere que as classes podem ser bem separadas com base na proximidade entre os dados, e o valor de K está ajustado corretamente.

- Se a Random Forest ou Regressão Logística tiverem desempenho melhor: Isso sugere que o problema tem interações complexas ou uma estrutura simples que pode ser modelada diretamente com essas abordagens.

# Conclusão:

- Modelos Simples (Logistic Regression) funcionam bem quando há uma relação linear clara entre as variáveis independentes e a classe.

- Modelos Complexos (Random Forest, SVM) podem ser mais eficazes quando o problema envolve interações não lineares ou padrões mais complexos.

- Modelos Baseados em Vizinhança (KNN) funcionam bem quando a separação entre as classes depende fortemente da proximidade entre os pontos
